```python
import csv, json
import pandas as pd
```

```python
data = []
with open('teams.csv','r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        data.append(row)
with open('teams.json','w') as jsonfile:
    json.dump(data,jsonfile,indent=2)
```
