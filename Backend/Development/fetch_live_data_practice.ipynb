```python
from nba_api.live.nba.endpoints import scoreboard, boxscore
from datetime import datetime, timezone
from dateutil import parser
f = "{gameId}: {awayTeam} vs. {homeTeam} @ {gameTimeLTZ}"

board = scoreboard.ScoreBoard()
print("ScoreBoardDate: " + board.score_board_date)
games = board.games.get_dict()
for game in games:
    gameTimeLTZ = parser.parse(game["gameTimeUTC"]).replace(tzinfo=timezone.utc).astimezone(tz=None)
    print(f.format(gameId=game,awayTeam=game['awayTeam']['teamName'],homeTeam=game['homeTeam']['teamName'],gameTimeLTZ=gameTimeLTZ))
```

```python
box = boxscore.BoxScore('0042400301')
box.game.get_dict()
```

```python
away_players_boxscore = box.away_team.get_dict()['players']
home_players_boxscore = box.home_team.get_dict()['players']
f="{player_id}: {name}: {points} PTS, {rebounds} rbs, {assists} AST"
print(f"==== {box.away_team.get_dict()['teamName']}: {box.away_team.get_dict()['score']} ====")
for player in away_players_boxscore:
    print(f.format(player_id=player['personId'],
                    name=player['name'],
                    points=player['statistics']['points'], 
                    rebounds=(player['statistics']['reboundsOffensive']+player['statistics']['reboundsDefensive']),
                    assists=player['statistics']['assists']))
print(f"==== {box.home_team.get_dict()['teamName']}: {box.home_team.get_dict()['score']} ====")
for player in home_players_boxscore:
    print(f.format(player_id=player['personId'],
                    name=player['name'],
                    points=player['statistics']['points'], 
                    rebounds=(player['statistics']['reboundsOffensive']+player['statistics']['reboundsDefensive']),
                    assists=player['statistics']['assists']))
```
